```python
import matplotlib.pyplot as plt
import numpy as np
from lab1lib import lab1lib as ll


DENSITY = 1000

VALS_PART = 0.005
WIDTH_PART = 0.3
CUT_COEFF = 1

STEP1 = 2
STEP2 = 2
STEP3 = 0.01

STEP = STEP1

FILE1 = "set_1.csv"
FILE2 = "set_2.csv"
FILE3 = "set_3.csv"

FILE = FILE1


def clear(distribution):
    return ll.clear(distribution, VALS_PART, WIDTH_PART, CUT_COEFF)
    #return distribution

def sort(distribution):
    return np.sort(distribution)


sorted_starting_distribution = sort(ll.read_csv(FILE))

size = len(sorted_starting_distribution)
expected_value = ll.get_expected_value(sorted_starting_distribution)
variance = ll.get_variance(sorted_starting_distribution, expected_value)
standart_deviation = variance ** 0.5

ll.draw_distribution_plot(sorted_starting_distribution, DENSITY)
```

```python
starting_cleared_sorted_distribution = clear(sorted_starting_distribution)

ll.draw_hist(starting_cleared_sorted_distribution, STEP)
```

```python
p = ll.get_geometric_p_by_expected_value(expected_value)
if p > 0 and p < 1:
    geometric_cleared_sorted_distribution_by_expected_value = clear(sort(np.random.geometric(p, size)))
    ll.draw_hist(geometric_cleared_sorted_distribution_by_expected_value, STEP)

#p = ll.get_geometric_p_by_variance(variance)
#if p > 0 and p < 1:
#    geometric_cleared_sorted_distribution_by_variance = clear(sort(np.random.geometric(p, size)))
#    ll.draw_hist(geometric_cleared_sorted_distribution_by_variance, STEP)
```

```python
p = ll.get_binomial_p_by_expected_value(expected_value, size)
binomial_cleared_sorted_distribution_by_expected_value = clear(sort(np.random.binomial(size, p, size)))
ll.draw_hist(binomial_cleared_sorted_distribution_by_expected_value, STEP)

#p = ll.get_binomial_p_by_variance(variance, size)
#binomial_cleared_sorted_distribution_by_variance = clear(sort(np.random.binomial(size, p, size)))
#ll.draw_hist(binomial_cleared_sorted_distribution_by_variance, STEP)
```

```python
poisson_cleared_sorted_distribution_by_expected_value = clear(sort(np.random.poisson(expected_value, size)))
ll.draw_hist(poisson_cleared_sorted_distribution_by_expected_value, STEP)

#poisson_cleared_sorted_distribution_by_variance = clear(sort(np.random.poisson(variance, size)))
#ll.draw_hist(poisson_cleared_sorted_distribution_by_variance, STEP)
```

```python
low, high = ll.get_uniform_low_high(expected_value, variance)
uniform_cleared_sorted_distribution = clear(sort(np.random.uniform(low, high, size)))

ll.draw_hist(uniform_cleared_sorted_distribution, STEP)
```

```python
normal_cleared_sorted_distribution = clear(sort(np.random.normal(expected_value, standart_deviation, size)))

ll.draw_hist(normal_cleared_sorted_distribution, STEP)
```

```python
exponential_cleared_sorted_distribution_by_expected_value = clear(sort(np.random.exponential(expected_value, size)))
ll.draw_hist(exponential_cleared_sorted_distribution_by_expected_value, STEP)

#exponential_cleared_sorted_distribution_by_variance = clear(sort(np.random.exponential(standart_deviation, size)))
#ll.draw_hist(exponential_cleared_sorted_distribution_by_variance, STEP)
```
